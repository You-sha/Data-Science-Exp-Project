## Plotting

Plotting is one of the **best ways to understand data**.

````python
cont_cols = df.select_dtypes(exclude=["object_"]).loc[:,:"days_posted_int"].drop('salary_mentioned',axis=1)

def hist(column):
    sns.histplot(cont_cols.iloc[:,column], edgecolor='black')

for i in np.arange(6):
    plt.rcParams["figure.figsize"] = (11,6)
    plt.subplots_adjust(wspace=0.8,hspace=0.5)
    plt.subplot(3,3,i+1)
    plt.suptitle('Continuous values', fontsize=20)
    hist(i)
    plt.ylabel('')
````

**Observations:**

* There seem to be a **lot** of outliers in ```reviews```.
* Jobs require **experience of around 5 years** on average.
* The majority of jobs were posted over a **month** ago.

___

Taking a look at how many jobs mention popular skills in their tags:

````python
yn_cols = df.select_dtypes(exclude=["object_"]).loc[:,'python':]

def plot(column):
    sns.set_context(rc = {'patch.linewidth': 0.5})
    sns.barplot(x=yn_cols.iloc[:,column].value_counts().index,y=yn_cols.iloc[:,column].value_counts(),edgecolor='black')

for i in np.arange(14):
    plt.rcParams["figure.figsize"] = (10,9)
    plt.subplots_adjust(wspace=0.8,hspace=0.8)
    plt.subplot(5,3,i+1)
    plt.title(f'{yn_cols.columns[i]}')
    plt.suptitle('Data Science skill requirements', fontsize=20)
    plot(i)
    plt.ylabel('')
````

**Observations:**

* Around **half of the jobs mention python**.
* Less than half of the jobs mention **SQL**.

___

Looking at jobs by **state**:

````python
df_locs = df.loc[:,'bangalore':'pune']

def plot1(column):
    sns.set_context(rc = {'patch.linewidth': 0.5})
    ax = sns.barplot(x=df_locs.iloc[:,column].value_counts().index,y=df_locs.iloc[:,column].value_counts(),edgecolor='black')

for i in np.arange(9):
    plt.rcParams["figure.figsize"] = (10,9)
    plt.subplots_adjust(wspace=0.8,hspace=0.8)
    plt.subplot(5,3,i+1)
    plt.suptitle('Data Science jobs by state', fontsize=20)
    plt.title(f'{df_locs.columns[i]}')
    plot1(i)
    plt.ylabel('')
````

**Observations:**

* **Most** of the jobs are located in **Bangalore**.
* **Kolkata** has the **lowest** amount of job postings.

___

Drawing some **pie charts** to understand the other variables:

````python
plt.rcParams["figure.figsize"] = (6,6)
plt.pie(df['salary_mentioned'].value_counts(), labels=['No','Yes'],autopct="%5.2f%%",\
        wedgeprops={"edgecolor":"k",'linewidth': 1})
plt.suptitle('Salary Mentioned',fontsize=20)
plt.show()
````

````python
plt.pie(df['job_simp'].value_counts(), labels=['Data Scientist','Na','MLE','Manager','Analyst','Data Engineer'],\
        autopct="%5.2f%%", wedgeprops={"edgecolor":"k",'linewidth': 0.1})
plt.suptitle('Job role',fontsize=20)
plt.show()
````

**Observations:**

* **The vast majority of jobs do not mention salary**. I wonder why that is?
* Around a 3rd of the jobs are for **senior positions**.
* Most of the jobs are asking for a **data scientist**. That makes sense, because I scraped the data with that as the keyword.

___

## Pivot Tables

To understand the data further, I am going to make some **pivot tables**:

````python
pd.pivot_table(df, index=['job_simp','post'], values=['bangalore', 'delhi', 'kolkata',
       'mumbai', 'remote', 'gurgaon', 'hyderabad', 'noida', 'pune'],aggfunc='sum').sort_values('bangalore', ascending=False)
````

**Observations:**

* **Bangalore** has the highest amount of postings for **data scientists**, and for **MLEs**.
* **Hyderabad** has the 2nd highest amount of job postings.

___


What is the **average experience** required for each **post**? And might as well have a look at **days posted**, **ratings** and **reviews**:

````python
pd.pivot_table(df, index=['job_simp','post'], values=['ratings','avg_experience','days_posted_int','reviews_int'])\
                                                      .sort_values('avg_experience',ascending=False)
````

**Observations:**

* A **senior manager** post requires the **most** amount of experience on average.
* A **senior data scientist** requires around **7 years** and a **junior** data scientist around **4 years**.
* All of the jobs were posted atleast over **22 days ago**.
* Companies hiring **analysts** have the **most reviews** .

___


Taking a look at **the number of each role** the **top 20 companies** are hiring:

````python
comp = df.company.value_counts().index[:20]
df['top_20_comp'] = df.company.apply(lambda x: x if x in comp else 'Na')
````

````python
pd.set_option('display.max_rows', 70)
pd.pivot_table(df, index=['top_20_comp','job_simp','post'],values=['avg_experience'], aggfunc='count').sort_values('avg_experience',ascending=False)
````

**Observation:** It is once again confirmed that **Skillety** is hiring the most.

# Salary Insights

Out of all the jobs that I have scraped, only **4.95% actually mention a salary**. I'm going to take a look at these rows now:

````python
df_sal = df.loc[df['salary_mentioned'] == 1].copy()

plt.rcParams["figure.figsize"] = (10,6)
sns.barplot(x=df_sal['salary'].value_counts().index, y=df_sal['salary'].value_counts())
plt.xticks(rotation=40)
plt.ylabel('')
plt.suptitle('Salary Range',fontsize=20)
plt.show()
````

````python
df_sal.salary = df_sal.salary.apply(lambda x: x.replace('50,000','0.5'))
df_sal['avg_salary'] = df_sal.salary.apply(lambda x: (float(x.split()[0].split('-')[0])+float(x.split()[0].split('-')[1]))/2)
df_sal.avg_salary.plot.hist()
plt.ylabel('')
plt.suptitle('Salary Distribution',fontsize=20)
plt.show()
````

**Observation:** Most companies offer a salary of around **10-15 LPA**.

___

Since this dataset is relatively small, I can see all the relevant information in just one pivot table:

````python
pd.pivot_table(df_sal,index=['location','company','job_simp','post'],values=['avg_salary','avg_experience', 'days_posted_int',
                                                                      'ratings','reviews_int'])
````

**Observations:**

* **None** of the top 20 companies have mentioned a salary.
* **TEG Analytics** offers the highest salary, **37.5 LPA**, for a **Permanent Remote, Senior Data Scientist** position.
* The **Permanent Remote** position has the most companies hiring.

# Word Cloud

Now, to see how many keywords are in the job's tags and what their frequency is, I am going to make a **word cloud**.

````python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
````

````python
words = " ".join(df['tags'])

def punctuation_stop(text):
    """remove punctuation and stop words"""
    filtered = []
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    for w in word_tokens:
        if w not in stop_words and w.isalpha():
            filtered.append(w.lower())
    return filtered


words_filtered = punctuation_stop(words)

text = " ".join([ele for ele in words_filtered])

wc= WordCloud(background_color="white", random_state=1,stopwords=STOPWORDS, max_words = 2000, width =800, height = 1500)
wc.generate(text)

plt.figure(figsize=[10,10])
plt.suptitle('Data Science Skills Wordcloud',fontsize=20)
plt.imshow(wc,interpolation="bilinear")
plt.axis('off')
plt.show()
````
